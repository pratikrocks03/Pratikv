```python
# Now Doing Text analysis on The_Belgian_Waffle_Co1 dataset
# Importing Libraies for Text Analysis
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
import string
from nltk.corpus import stopwords
```

```python
df=The_Belgian_Waffle_Co1
```

```python
df.head()
```

```python
df.shape
```

136 rows and 2 columns present in the dataset.

```python
# new columns for the length of Item Description
df['length']=df.Item_Description.str.len()
df.head()
```

Now we can see one column 'length' has been added in the datset and giving length of the Item description.

```python
# converting all item name and item description to lower case
df['Item_Name']=df['Item_Name'].str.lower()
df['Item_Description']=df['Item_Description'].str.lower()
                   
```

```python
 # Replace money symbols with 'moneysymb'
df['Item_Name']=df['Item_Name'].str.replace(r'£|\$','dollers')

df['Item_Description']=df['Item_Description'].str.replace(r'£|\$','dollers')
```

```python
# Replace 10 digit phone numbers(formats include paranthesis, spaces, no spaces, dashes) with 'phonenumber'
df['Item_Name']=df['Item_Name'].str.replace(r'^\(?[\d]{3}\)?[\s-]?[\d]{3}[\s-]?[\d]{d}','phonenumber')

df['Item_Description']=df['Item_Description'].str.replace(r'^\(?[\d]{3}\)?[\s-]?[\d]{3}[\s-]?[\d]{d}','phonenumber')
```

```python
# replace numbers with 'numbr'
df['Item_Name']=df['Item_Name'].str.replace(r'\d+(\.\d+)?','numbr')

df['Item_Description']=df['Item_Description'].str.replace(r'\d+(\.\d+)?','numbr')

# remove punctuation
df['Item_Name']=df['Item_Name'].str.replace(r'[^\w\d\s]',' ')

df['Item_Description']=df['Item_Description'].str.replace(r'[^\w\d\s]',' ')


# Replace whitespace between terms with a single space
df['Item_Name']=df['Item_Name'].str.replace(r'\s+',' ')

df['Item_Description']=df['Item_Description'].str.replace(r'\s+',' ')

# Remove Leading and trailing whitespace
df['Item_Name']=df['Item_Name'].str.replace(r'^\s+|\s+?$',' ')

df['Item_Description']=df['Item_Description'].str.replace(r'^\s+|\s+?$',' ')
```

```python
# Removing stop words

stop_words=set(stopwords.words('english')+['u','ur','4','2','im','dont','doin','ure'])

df['Item_Name']=df['Item_Name'].apply(lambda x:' '.join(term for term in x.split() if term not in stop_words))

df['Item_Description']=df['Item_Description'].apply(lambda x: ' '.join(term for term in x.split() if term not in stop_words))
```

```python
# New column (clean_length) after punctuations, stop words removal
df['clean_length']=df.Item_Description.str.len()
df.head()
```

```python
# Total Length removal
print('original length',df.length.sum())
print('clean length',df.clean_length.sum())
```

```python
# Getting sense of Loud words Item_Name

from wordcloud import WordCloud

loud=df['Item_Name']

loud_cloud=WordCloud(width=700,height=500,background_color='white',max_words=500).generate(' '.join(loud))
print('Item Name')
plt.figure(figsize=(10,8),facecolor='r')
plt.imshow(loud_cloud)
plt.axis('off')
plt.tight_layout(pad=0)
plt.show()
```

```python
# Getting sense of Loud words Item_Description

from wordcloud import WordCloud

loud=df['Item_Description']

loud_cloud=WordCloud(width=700,height=500,background_color='white',max_words=500).generate(' '.join(loud))
print('Item Description')
plt.figure(figsize=(10,8),facecolor='r')
plt.imshow(loud_cloud)
plt.axis('off')
plt.tight_layout(pad=0)
plt.show()
```

We can see Bold words and other words are indication of Loud words that is most number of used words in Item Name and Item Description columns.
This analysis os very important for comparing from other restaurents.
