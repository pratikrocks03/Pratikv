```python
allpn=[]
allasn=[]
allid=[]
alln=[]
alld=[]
allt=[]
alls=[]
allr=[]
allc=[]
Asin=[]
for i in soup:
       for d in i.findAll('div', attrs={'class':"a-section review aok-relative"}):
            ID=d["id"]
            ASIN=d.find('a',attrs={'class':"a-text-bold"})
            productname=i.find('a',attrs={'class':"a-link-normal"})
            name = d.find('span', attrs={'class':'a-profile-name'})
            Date = d.find('span', attrs={'class':'a-size-base a-color-secondary review-date'})
            Title = d.find('a', attrs={'class':"a-size-base a-link-normal review-title a-color-base review-title-content a-text-bold"})
            Product_speci = d.find('div', attrs={"a-row a-spacing-mini review-data review-format-strip"})
            Rating = d.find('span', attrs={'class':"a-icon-alt"})
            comment = d.find('span', attrs={'class':"a-size-base review-text review-text-content"})
            #print(Title)
            #print(productname)
            #print(ID)
            
            if ID is not None:
                allid.append(ID)
            else:
                allid.append("unknown-ID")
            if Asin is not None:
                allasn.append(ASIN)
            else:
                allasn.append("unknown-Asin")
            if productname is not None:
                allpn.append(productname.text)
            else:
                allpn.append("unknown-product")
            if name is not None:
                alln.append(name.text)
            else:
                alln.append("unknown-product")
            if Date is not None:
                alld.append(Date.text)
            else:
                alld.append("unknown-product")
            if Title is not None:
                allt.append(Title.text)
            else:
                allt.append("unknown-product")
            if Product_speci is not None:
                alls.append(Product_speci.text)
            else:
                alls.append("unknown-product")
            if comment is not None:
                allc.append(comment.text)
            else:
                allc.append("unknown-product")
            if Rating is not None:
                allr.append(Rating.text)
            else:
                allr.append("unknown-product")
   
```

```python
a={'ID':allid,'ASIN':allasn,'Product Name':allpn,'Name':alln, 'Date':alld,'Title':allt,'Specification':alls,'Rating':allr,'comment':allc}
output=pd.DataFrame.from_dict(a, orient='index')
output=output.transpose()
```
