# Importing Libraries

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import pandas as pd
```

# Scrapping Process

```python
# Set up the WebDriver
options = webdriver.ChromeOptions()
options.add_argument("--headless")  # Run browser in headless mode (optional)
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

# Navigate to the Amazon product page
product_url = "https://www.amazon.in/dp/B07K6L3KC6"  # Replace with actual product URL or ASIN
driver.get(product_url)

wait = WebDriverWait(driver, 10)
reviews_section = wait.until(EC.presence_of_element_located((By.ID, "reviewsMedley")))

reviews = []
product_name = driver.find_element(By.ID, "productTitle").text  # Get the product name

while True:
    review_elements = driver.find_elements(By.CSS_SELECTOR, ".review")

    for review_element in review_elements:
        review_id = review_element.get_attribute("id")  # Get the review ID
        person_name = review_element.find_element(By.CSS_SELECTOR, ".a-profile-name").text
        date = review_element.find_element(By.CSS_SELECTOR, ".review-date").text
        title = review_element.find_element(By.CSS_SELECTOR, ".review-title-content").text
        rating = review_element.find_element(By.CSS_SELECTOR, ".review-rating").get_attribute("innerText")
        comment = review_element.find_element(By.CSS_SELECTOR, ".review-text-content").text

        reviews.append({
            "product_name": product_name,
            "review_id": review_id,
            "person_name": person_name,
            "date": date,
            "title": title,
            "rating": rating,
            "comment": comment
        })
    
    # Check if there is a next page
    try:
        next_button = driver.find_element(By.CSS_SELECTOR, ".a-last a")
        next_button.click()
        time.sleep(2)  # Wait for the next page to load
    except Exception as e:
        print("No more pages or error:", e)
        break
```
